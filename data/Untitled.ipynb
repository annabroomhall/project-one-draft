```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import json as json

# File to Load (Remember to Change These)
crime_2017 = "2017-18-data_sa_crime.csv"
crime_2018 = "2018-19-data_sa_crime.csv"
crime_2019 = "2019-20-fullyr-data_sa_crime (1).csv"
crime_2020 = "2020-21_crime_qtr1_and_qtr2 (1).csv"
sa_postcodes = "australian_postcodes (1).csv"

# Read School and Student Data File and store into Pandas DataFrames
crime_2017_df = pd.read_csv(crime_2017)
crime_2018_df = pd.read_csv(crime_2018)
crime_2019_df = pd.read_csv(crime_2019)
crime_2020_df = pd.read_csv(crime_2020)
sa_postcodes_df = pd.read_csv(sa_postcodes)
```

```python
merged_data = pd.concat([crime_2017_df, crime_2018_df,crime_2019_df,crime_2020_df], ignore_index=True)
```

```python
clean_crime_data = merged_data.dropna(how='any')
```

```python
clean_crime_data.dtypes
```

```python
sa_postcodes_df = sa_postcodes_df.rename(columns = {"postcode":"Postcode - Incident"})
```

```python
clean_crime_data = clean_crime_data.loc[clean_crime_data["Postcode - Incident"]!= "NOT DISCLOSED"].copy()
```

```python
clean_crime_data["Postcode - Incident"] = clean_crime_data["Postcode - Incident"].astype(str).astype(int)
```

```python
clean_crime_data.dtypes
```

```python
sa_postcodes_df.dtypes
```

```python
crime_postcode_merge_df = pd.merge(clean_crime_data, sa_postcodes_df, on = "Postcode - Incident", how = "left")
```
